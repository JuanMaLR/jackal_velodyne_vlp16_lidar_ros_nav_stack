```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_tf = []
y_tf = []
substrings = []
with open("/content/drive/MyDrive/Thesis/Transformations txt/Test_1_2D_LiDAR.txt") as fh:
    for line in fh:
        if line.startswith("- Translation:"):
          split_str = line.split("[")

          for s in split_str[1:]:
              split_s = s.split("]")
              if len(split_s) > 1:
                  substrings.append(split_s[0])

          x_tf.append(substrings[0].split(',')[0])
          y_tf.append(substrings[0].split(',')[1])

          substrings.clear() 
```

```python
x_tf_int = [eval(i) for i in x_tf]
y_tf_int = [eval(i) for i in y_tf]
```

```python
plt.figure(figsize=(16,8));
plt.plot(y_tf_int, x_tf_int, label="Robot's trajectory")
#Set title
plt.title('2D LiDAR Test 1');
#Set x and y axes labels
plt.xlabel("Map's reference frame y-axis");
plt.ylabel("Map's reference frame x-axis");
#Display legend
plt.legend(loc="upper left")
#Reduce step size in y axis ticks
#plt.yticks(np.arange(min(y_x), max(y_y)+1, 2));
#Show grid
plt.grid()
plt.axis('equal')
```
